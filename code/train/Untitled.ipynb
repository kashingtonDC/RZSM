```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
files = [x for x in os.listdir(os.getcwd())]
```

```python
dfs = []
for i in files:
    if i.endswith(".txt"):
        dfs.append(pd.read_csv(i))
    else:
        continue
```

```python
t = pd.concat(dfs)
```

```python
t2=t.dropna()
```

```python
q = t2.rzsm.quantile(0.99)
```

```python
t2 = t2[t2.rzsm < q]
```

```python
plt.scatter(t2.rzsm, t2.sigma)
```
